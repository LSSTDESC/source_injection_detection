## Set globals

```python
SYSTEM_INDEX_LIST = [1, 4] #[1, 2, 4, 5]
TIME_INDEX_LEN = 5
FOLDER = tl.STAMP_FOLDER

os.makedirs(FOLDER, exist_ok=True)
```

## Single stamps

Run through systems. Skip #3, #6 system because the size is 36x36 not 30x30. 
Note #2 system has arc flux < 0 (why?); #5 system also shows this sometimes.
So focus on #1 and #4. 

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)
    
    for TIME_INDEX in range(TIME_INDEX_LEN):

        print('-'*20)
        print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}, TIME_INDEX {TIME_INDEX}')
        
        stp.get_single_stamp(SYSTEM_INDEX, TIME_INDEX, FOLDER)
```

## Coadd stamps

Go through systems for coadd. 

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)

    print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}')
        
    stp.get_coadd_stamp(SYSTEM_INDEX, FOLDER)
```

## WCS

Add wcs (for injection and rotation).

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)
    filename = "%s/system_%d_coadd.fits"%(FOLDER, SYSTEM_INDEX)
    stp.add_wcs(filename)
    
    for TIME_INDEX in range(TIME_INDEX_LEN):

        print('-'*20)
        print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}, TIME_INDEX {TIME_INDEX}')

        filename = "%s/system_%d_%d.fits"%(FOLDER, SYSTEM_INDEX, TIME_INDEX)
        stp.add_wcs(filename)

```
