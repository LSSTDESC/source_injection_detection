## Set globals

```python
SYSTEM_INDEX_LIST = range(14)
#TIME_INDEX_LEN = 7
TIME_INDEX_LEN = 16
#TIME_INDEX_LEN = 18
FOLDER = tl.STAMP_FOLDER
BAND = tl.BAND

os.makedirs(FOLDER, exist_ok=True)
```

## Single stamps

Run through systems. 

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)
    
    for TIME_INDEX in range(TIME_INDEX_LEN):

        print('-'*20)
        print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}, TIME_INDEX {TIME_INDEX}')
        
        stp.get_single_stamp(SYSTEM_INDEX, TIME_INDEX, BAND, FOLDER)
```

## Coadd stamps

Go through systems for coadd. 

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)

    print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}')
        
    stp.get_coadd_stamp(SYSTEM_INDEX, BAND, FOLDER)
```

## WCS

Add wcs (for injection and rotation).

```python
for SYSTEM_INDEX in SYSTEM_INDEX_LIST:
    
    print('='*30)
    filename = f"{FOLDER}/system_{SYSTEM_INDEX}_coadd_{BAND}.fits"
    stp.add_wcs(filename)
    
    for TIME_INDEX in range(TIME_INDEX_LEN):

        print('-'*20)
        print(f'\nSYSTEM_INDEX {SYSTEM_INDEX}, TIME_INDEX {TIME_INDEX}')

        filename = f"{FOLDER}/system_{SYSTEM_INDEX}_{TIME_INDEX}_{BAND}.fits"
        stp.add_wcs(filename)

```
