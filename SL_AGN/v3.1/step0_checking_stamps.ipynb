## Set globals

```python
LENS_INDEX_MIN = tl.LENS_INDEX_MIN    
N_LENS = tl.N_LENS
print(f"LENS_INDEX_MIN: {LENS_INDEX_MIN}")
print(f"N_LENS: {N_LENS}")

FOLDER = tl.STAMP_FOLDER
BAND = tl.BAND
print(f"FOLDER: {FOLDER}")
print(f"BAND: {BAND}")

# Cap at total number of systems in the H5 file
import h5py
with h5py.File(stp.LENS_FILENAMES[0], 'r') as h5f:
    N_TOTAL = len(h5f.keys())
LENS_INDEX_MAX = min(LENS_INDEX_MIN + N_LENS, N_TOTAL)
print(f"N_TOTAL: {N_TOTAL}, LENS_INDEX_MAX: {LENS_INDEX_MAX}")

Path(FOLDER).mkdir(parents=True, exist_ok=True)
```

## Static stamps

Save one static image (lens galaxy + host galaxy / arc) per system.

```python
for i in range(LENS_INDEX_MIN, LENS_INDEX_MAX):
    system_index = i
    if system_index % 100 == 0:
        print(f"system_index: {system_index}")
    stp.get_static_stamp(system_index, BAND, FOLDER)
```

## WCS

Add wcs (for injection and rotation).

```python
for i in range(LENS_INDEX_MIN, LENS_INDEX_MAX):
    system_index = i
    if system_index % 100 == 0:
        print(f"system_index: {system_index}")

    filename = f"{FOLDER}/system_{system_index}_static_{BAND}.fits"
    p = Path(filename)
    if p.exists():
        stp.add_wcs(filename)
        p.unlink()
```
