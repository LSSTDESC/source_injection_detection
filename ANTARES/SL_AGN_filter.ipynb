## Define filter

Schema: https://antares.noirlab.edu/properties

```python
class LensedAGN(dk.Filter):
    NAME = "Lensed_AGN"
    ERROR_SLACK_CHANNEL = None
    REQUIRED_LOCUS_PROPERTIES = [
        'ztf_object_id',
        #'locus_id',
    ]
    REQUIRED_ALERT_PROPERTIES = [
        'ztf_magdiff',
        'ztf_seeratio',
        'ztf_rb',
    ]
    OUTPUT_LOCUS_PROPERTIES = []
    OUTPUT_ALERT_PROPERTIES = []
    OUTPUT_TAGS = [
        {
            'name': 'lensed_agn',
            'description': 'An alert that shows signatures of coming from a lensed AGN.',
        },
    ]
    def run(self, locus):
        locus_id = locus.locus_id
        print('Locus id: ', locus_id)
        #locus = dk.get_locus(locus_id)
        locus_dict = locus.to_dict()
        #print('locus_dict: ', locus_dict)
        alerts = locus_dict['alerts']
        #print('alert: ', alerts[0])
        for count, alert in enumerate(alerts):
            if count==10: break
            alert_properties = alert['properties']
            col = 'ztf_magdiff'
            if col in alert_properties.keys():
                value = alert_properties[col]
                if value < 0:
                    print('Find one candidate!')
                    #locus.tag('Lensed AGN candidate')
                    return 0
                    
```

## Run filter

https://sled.amnh.org/lenses/detail/3449

```python
locus_id = 'ANT2020ts33s'
report = dk.run_filter(LensedAGN, locus=locus_id)
print(report)
```
