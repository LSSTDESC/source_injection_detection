## Initialize

Run on NOIRLab Data Lab. Choose ANTARES Python 3 Kernel.

```python
import antares.devkit as dk
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

dk.init()
```

## Define filter

Schema: https://antares.noirlab.edu/properties

```python
class LAGN(dk.Filter):
    OUTPUT_TAGS = [
        {
            'name': 'lagn',
            'description': 'lagn',
        },
    ]

    def run(self, locus):
        
        magdiff_threshold = -0.1
        
        print('Locus: ', locus.locus_id)
        
        alert_magdiff = locus.alert.properties['ztf_magdiff']

        if alert_magdiff < magdiff_threshold:
            print("Find one lagn candidate!")
            locus.tag('lagn')
```

## Run filter

https://sled.amnh.org/lenses/detail/3449

```python
locus_id = 'ANT2020ts33s'
report = dk.run_filter(LAGN, locus=locus_id)
print(report)
```
